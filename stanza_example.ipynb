```python
import stanza
```

```python
stanza_lat = stanza.Pipeline('la', processors='tokenize,lemma,pos,depparse')
```

```python
stanza_results = stanza_lat("Lorem ipsum dolor sit amet, consectetur adipisci velit. ")
for sent in stanza_results.sentences:
  for word in sent.words:
    print(word.text)
    print(word.lemma)
    print(word.upos)
    print(word.feats)
    print("XXXXXXXXXXXX")
```
